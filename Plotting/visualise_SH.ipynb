```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from plots import plot_SH, plot_exp_sin, plot_tex


%load_ext autoreload
%autoreload 2
```

```python
fig = plt.figure(figsize=plt.figaspect(1.))
ax = fig.add_subplot(projection='3d')
l, m = 4, 0
plot_SH(ax, l, m)
plt.savefig('Y{}_{}.svg'.format(l, m), format='svg')
plt.show()
```

```python
ls = np.arange(0,7,2) # l=  0, 2, 4, 6
figsize_px, DPI = 300, 400
figsize_in = figsize_px / DPI
fig = plt.figure(figsize=(16, 2), dpi=DPI)
spec = gridspec.GridSpec(ncols=2*ls[-1]+1, nrows=1, figure=fig)
for i, l in enumerate(ls):
    ax = fig.add_subplot(spec[0, i], projection='3d')
    plot_SH(ax, l)
#plt.tight_layout()
plt.savefig('sph_harm_samples.svg', format='svg')
#plt.text(x=1200,y=1,s="Spherical Harmonics")
plt.show()
```

```python
fig = plt.figure(figsize=plt.figaspect(1.))
ax = fig.add_subplot(projection='3d')
A, B = 1, 0.5
plot_exp_sin(ax, A, B)
plt.savefig('F{}_{}.svg'.format(A, B), format='svg')
plt.show()
```

```python
ls = np.arange(2,15,4) # l=  0, 2, 4, 6
figsize_px, DPI = 800, 400
figsize_in = figsize_px / DPI
fig = plt.figure(figsize=(36, 2), dpi=DPI)
spec = gridspec.GridSpec(ncols=2*ls[-1]+1, nrows=1, figure=fig)
for i, l in enumerate(ls):
    ax = fig.add_subplot(spec[0, i], projection='3d')
    plot_exp_sin(ax, A=1, B = l)
#plt.tight_layout()
plt.savefig('exp_sin_samples.svg', format='svg')
#plt.text(x=1200,y=1,s="Spherical Harmonics")
plt.show()
```
